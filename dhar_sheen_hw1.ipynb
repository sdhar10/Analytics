```python
"""1. Assume that you are responsible for counting the number of students who are graduating from Purdue this term based on their responses to a survey. A student who is graduating may submit their response as ‘G’ or ‘g’ and a student who is not graduating could either answer as ‘N’ or ‘n’. These responses are stored in a list called status """

status = "g G n n n N g G g N G G N g n"
num_grad = 0
num_not_grad = 0
total = 0 

for i in status :
    
    if i == 'G' or i == 'g':
        num_grad+=1
    
    elif i == 'N' or i == 'n':
        num_not_grad+=1

status = status.replace(" ","")
total = len(status)

print('Total % of students graduating :' , (num_grad/total)*100)
print('Total % of students not graduating :' , (num_not_grad/total)*100)
```

```python
"""2. Pratyush has a list of items (Harry Potter, Gone Girl, etc.) he wants to complete this Summer and has stored the genres of each item (Fantasy Fiction, Mystery, etc.), in several dictionaries based on the type of activity (Books, Movies or Video games)."""
books = {'Harry Potter': 'Fantasy Fiction', 'Gone Girl': 'Mystery'}
movies = dict([('Seabiscuit','Drama/Sport'), ('House of Wax', 'Horror/Thriller')])
video_games = dict(StardewValley='Adventure', Fortnite='Survival')

print('\n')
print('Pratyushs summer items:')
print(list(books.keys()))
print(list(movies.keys()))
print(list(video_games.keys()))


print('\n')
print('All the genres:')
print(list(books.values()))
print(list(movies.values()))
print(list(video_games.values()))

print('\n')
print('As pairs:')
print(list(books.items()))
print(list(movies.items()))
print(list(video_games.items()))

print('\n')
print('Remove Gone Girl:')
del books ['Gone Girl']
print(list(books.keys()))

print('\n')
print('A single dictionary:')
dest = dict(books)
dest.update(movies)
dest.update(video_games)
print(dest)

print('\n')
print('Add Cruella:')
dest['Cruella'] = 'Comedy'
print(dest)

print('\n')
print('Genre for House of Wax?:')
print(dest.get('House of Wax'))

print('\n')
print('Remove everything:')
dest.clear()
print(dest)
```

```python
"""Create a Python list with the following values: [3, 5, 9, 20, 30, 55, 102, 123, 174]"""

agg = [3, 5, 9, 20, 30, 55, 102, 123, 174]
summation = 0
prod = 1 
print( 'Fifth element of the list :', agg[4])

for i in agg:
    summation += i
print('Sum of the values in the list :', summation)

print('Mean of the values in the list :', summation/len(agg))

for i in agg:
    prod *= i
print('Sum of the values in the list :', prod)

    
```

```python
"""FizzBuzz is a classic programming exercise. Iterate over the numbers from 0 to 99 and for each number"""
fizz = []
buzz = []
fizzbuzz = []

for i in range(100):
    if i%3 == 0 and i%5!=0:
        print(i ,'Fizz')
        fizz.append(i)
    elif i%5 == 0 and i%3!=0: 
        print(i,'Buzz')
        buzz.append(i)
    elif i%3 == 0 and i%5==0:
        print(i,'Fizz Buzz')
        fizzbuzz.append(i)
        
print('Numbers divisible by ONLY 3:', fizz)
print('Numbers divisible by ONLY 5:', buzz)
print('Numbers divisible by 3 & 5:', fizzbuzz)

```
